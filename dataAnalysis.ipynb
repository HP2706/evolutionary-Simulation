```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_parquet("test.parquet")


```

```python
# Unique agent IDs
agent_ids = df['agent_id'].unique()

# Setting up the figure size for better visibility

for y_axis in ['score', 'count', 'fitness']:
    plt.figure(figsize=(15, 10))
    plt.title(f"{y_axis}")
    plt.xlabel("Round")
    plt.ylabel(y_axis)
    for agent_id in agent_ids:
        agent_df = df[df['agent_id'] == agent_id]
        plt.plot(agent_df['round_number'], agent_df[y_axis], label=f"Agent {agent_id}")
    plt.legend()
    plt.show()
```
